```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
sm = pd.read_csv('g://sm.csv')
sm
```

```python
sb.stripplot(x='OS',y='Ram',data=sm,size=10, jitter=True)
plt.show()
```

```python
plt.figure(figsize=(11,6))
sb.swarmplot(x='OS', y='Capacity', data=sm, hue='Ram', size=10)
plt.show()
#swarmplot has not overlap
```

```python
plt.figure(figsize=(11,6))
sb.boxplot(x='Company', y='Ram', data=sm)
plt.show()
```

```python
sb.jointplot(x='Capacity', y='Ram', data=sm,kind='scatter')
plt.show() 
```

```python
sb.set_style('darkgrid')
sb.pairplot(sm, hue='Name', plot_kws={'s':90})
plt.show()
```
